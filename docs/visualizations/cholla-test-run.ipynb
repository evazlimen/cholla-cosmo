Notebook to plot the results of a simple test calculation on lux

```python
#import necessary packages
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
#set the output that you want to look at
num = 1
```

```python
#set the path to your data directory
fdir = f"/data/groups/comp-astro/ezlimen/cosmo-cholla-tests/run/data/{num}/"
```

First, let's look at the hydro output:

```python
#set the path to the hydro output file
fname = fdir + f'{num}.h5'
```

```python
#load the hdf5 file
source = h5py.File(fname,'r')
```

```python
#As an aside, here are the keys you can use to access data within the hdf5 file:
list(source.keys())
```

```python
#Let's make a quick density plot
#First, we need to get the density data
x = source['density']
```

```python
#Let's see what we are working with:
x.shape
```

```python
#To do a 2-D visualization, we can make a projection along the z-axis
xproj = np.sum(x,axis=2)
```

```python
#Now let's plot it!
plt.figure(figsize=(4,4))
plt.imshow(np.log10(xproj), origin = "lower", vmin =2.5, vmax = 4.75) #set the origin of the image to the lower left corner

# set colorbar
cbar = plt.colorbar(fraction = 0.046, pad = 0.04) #resize the colorbar so it looks nicer
cbar_lbl = "Density"
plt.xlabel("X (cells)")
plt.ylabel("Y (cells)")
cbar.ax.set_ylabel(cbar_lbl, rotation=270, labelpad=15)
plt.title("Hydro density")
plt.tight_layout()
```

Now, let's look at the particle output:

```python
#set the path to the particle output file
fname = fdir + f'{num}_particles.h5'
```

```python
#load the hdf5 file
part_source = h5py.File(fname,'r')
```

```python
#Again, let's see what keys we can use to access information
list(part_source.keys())
```

```python
#Let's make another quick density plot
#First, we need to get the density data
rho = part_source['density']
```

```python
#Let's see what we are working with:
rho.shape
```

```python
#Now let's plot it!
plt.figure(figsize=(4,4))
plt.imshow(np.log10(rhoproj), origin = "lower", vmin =2.5, vmax = 4.75) #set the origin of the image to the lower left corner

# set colorbar
cbar = plt.colorbar(fraction = 0.046, pad = 0.04) #resize the colorbar so it looks nicer
cbar_lbl = "Density"
plt.xlabel("X (cells)")
plt.ylabel("Y (cells)")
cbar.ax.set_ylabel(cbar_lbl, rotation=270, labelpad=15)
plt.title("Particle density")

plt.tight_layout()
```
